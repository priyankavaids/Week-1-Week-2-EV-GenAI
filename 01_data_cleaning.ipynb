```python
import pandas as pd

# Load your uploaded file
df = pd.read_csv("ev_uploaded_copy.csv")

# Display basic info
print("Shape of dataset:", df.shape)
df.head()
```

```python
df.columns = [c.strip().lower().replace(' ', '_').replace('-', '_') for c in df.columns]
df.head()
```

```python
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"Removed {before - after} duplicate rows.")
```

```python
df.isnull().sum()
```

```python
# Separate numeric and text columns
num_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Fill missing numeric with median, text with 'Unknown'
for c in num_cols:
    df[c] = df[c].fillna(df[c].median())

for c in cat_cols:
    df[c] = df[c].fillna('Unknown')

print("Missing values filled successfully!")
```

```python
cleaned_filename = "ev_specs_cleaned.csv"
df.to_csv(cleaned_filename, index=False)

from google.colab import files
files.download(cleaned_filename)
```
